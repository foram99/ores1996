```python
# importing pandas, numpy,scipy
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#renaming few Columns
df = df.rename(columns={"High School Diploma": "High_School", "With a disability" : "Disabled"})
```

```python
#droping duplicate rows
df = df.drop_duplicates()
```

```python
#droping columns
drop_columns = ['Spanish Spoken in HH', 'Chinese Spoken in HH', 'Korean Spoken in HH', 'Vietnamese Spoken in HH', 'Tagalog Spoken in HH', 'Russian Spoken in HH']

# Useing drop method to remove the specified columns
df = df.drop(columns=drop_columns)
```

```python
#checking columns after droping few previously
df.columns
```

```python
# Iterate through all columns and replace "**" with "0"
for column in df.columns:
    df[column] = df[column].str.replace(r'\*\*', '0')
df
```

```python
#replacing the missing value
df = df.fillna('')
df
```

```python
# Strip leading and trailing whitespace and convert to capital letters
df['State Name'] = df['State Name'].str.strip().str.upper()
df
```

```python
# Deleting rows containing 'USTOTAL'
df = df[df.ne('US TOTAL').all(axis=1)]

# Resetting the index
df.reset_index(drop=True, inplace=True)
df
```

```python
#Remove leading whitespace and update the DataFrame
df['Non-Elderly Unisured Population'] = df['Non-Elderly Uninsured Population'].str.lstrip()
```

```python
# coverting data from 'object' to 'float'
columns_to_convert = [
    "Non-Elderly Population",
    "Non-Elderly Uninsured Population",
    "HIU Income < 100% FPL",
    "HIU Income 100-138% FPL",
    "HIU Income 139-249% FPL",
    "HIU Income 250-400% FPL",
    "HIU Income 400% FPL",
    "Age 0-18",
    "Age 19-34",
    "Age 35-49",
    "Age 50-64",
    "Male",
    "Female",
    "Less than High School",
    "High_School",
    "College Grad",
    "Married",
    "Got married in last 12 months",
    "Gave birth in last 12 months",
    "Child in Family",
    "Spanish/Hispanic/Latino Origin",
    "White Non-Latino",
    "Black Non-Latino",
    "Asian / Native-Hawaiian / Pac Islander",
    "American Indian / Alaska Native",
    "Multi-racial or Other",
    "Internet Access in HH",
    "Cellular data plan for mobile device in HH",
    "High Speed Internet service in HH",
    "SNAP Recipient",
    "Disabled",
    "Full-time Worker in Family",
    "Self-Employed, Incorporated",
    "Self-Employed, Unincorporated",
    "Employed in Agriculture Industry",
    "Employed in Mining/Construction Industry",
    "Employed in Manufacturing Industry",
    "Employed in Trade Industry",
    "Employed in Info/Finance Industry",
    "Employed in Education/Health Industry",
    "Employed in Entertainment Industry",
    "Employed in Service Industry",
    "Employed in Military/Public Industry",
    "No English Speaking Adults in HH",
    "English Speaker",
    "Other Language Spoken in HH"
]

for column in columns_to_convert:
    if df[column].dtype == 'object':
        try:
            df[column] = df[column].str.replace(',', '').astype(float)
        except ValueError as e:
            print(f"Error converting column '{column}': {e}")

df["Non-Elderly Percent Uninsured"] = df["Non-Elderly Percent Uninsured"].str.replace('%', '').astype(float)

```

```python
# Create the calculated field "Total Children" by combining two columns
df["Total Children"] = df["Gave birth in last 12 months"] + df["Child in Family"]
```

```python
# Create the calculated field "total married" by combining two columns
df["Total married"] = df["Married"] + df["Got married in last 12 months"]
df
```

```python
# Create the calculated field "Total Married"
df["Industrial Employed"] = df["Employed in Agriculture Industry"] + df["Employed in Mining/Construction Industry"] + df["Employed in Manufacturing Industry"] + df["Employed in Trade Industry"] + df["Employed in Info/Finance Industry"] + df["Employed in Education/Health Industry"] + df["Employed in Entertainment Industry"] + df["Employed in Service Industry"] + df["Employed in Military/Public Industry"]

# Print the "Industrial Employed" column
df

```

```python
columns_to_drop = [
    'Married',
    'Got married in last 12 months',
    'Gave birth in last 12 months',
    'Child in Family',
    'Employed in Agriculture Industry',
    'Employed in Mining/Construction Industry',
    'Employed in Manufacturing Industry',
    'Employed in Trade Industry',
    'Employed in Info/Finance Industry',
    'Employed in Education/Health Industry',
    'Employed in Entertainment Industry',
    'Employed in Service Industry',
    'Employed in Military/Public Industry'
]

# Use the drop method to remove the specified columns
df = df.drop(columns=columns_to_drop)
```

```python
df.info()
```

```python
# Setting threshold for identifying outliers
threshold = 3

# Creating new DataFrame to store outlier flags
outlier_flags = pd.DataFrame()

# Calculating Z-scores and identify outliers for each numeric column
for column in df.select_dtypes(include=[np.number]).columns:
    z_scores = np.abs(stats.zscore(df[column]))
    outliers = np.abs(z_scores) > threshold
    outlier_flags[column] = outliers
```

```python
# Filter rows with outliers and remove them
filtered_df = df[~outlier_flags.any(axis=1)]
df
```

```python
from sklearn.preprocessing import StandardScaler

# Separating the numeric and non-numeric columns
numeric_columns = filtered_df.select_dtypes(include=[int, float]).columns
non_numeric_columns = filtered_df.select_dtypes(exclude=[int, float]).columns

#StandardScaler
scaler = StandardScaler()

# Fitting the scaler to the numeric columns and transforming them
filtered_df[numeric_columns] = scaler.fit_transform(filtered_df[numeric_columns])

# Printting  the DataFrame with standardized numeric columns
print("DataFrame with Standardized Numeric Columns:")
print(filtered_df)
```

```python
# Specify the file path where it is saved in the CSV file format
file_path = "/content/drive/MyDrive/cleaned_dataset.csv"

# Save the DataFrame to a CSV file
filtered_df.to_csv(file_path, index=False)

print(f"DataFrame saved to {file_path}")
```
